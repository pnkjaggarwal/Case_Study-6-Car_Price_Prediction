```python
car['doornumber'].replace({'two':2, 'four':4}, inplace=True)
car['cylindernumber'].replace({'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8}, inplace=True)
```

```python
car = pd.get_dummies(car, drop_first=True)
```

```python
X = car.drop('price', axis=1)
y = car['price']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=0)
```

```python
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

```python
scale = StandardScaler()
```

```python
x_train.columns
```

```python
x_train[['symboling', 'doornumber', 'wheelbase', 'carlength', 'carwidth',
       'carheight', 'curbweight', 'cylindernumber', 'enginesize', 'boreratio',
       'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg',
       'highwaympg']] = scale.fit_transform(x_train[['symboling', 'doornumber', 'wheelbase', 'carlength', 'carwidth',
       'carheight', 'curbweight', 'cylindernumber', 'enginesize', 'boreratio',
       'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg',
       'highwaympg']])
x_train.head()
```

```python
x_test[['symboling', 'doornumber', 'wheelbase', 'carlength', 'carwidth',
       'carheight', 'curbweight', 'cylindernumber', 'enginesize', 'boreratio',
       'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg',
       'highwaympg']] = scale.transform(x_test[['symboling', 'doornumber', 'wheelbase', 'carlength', 'carwidth',
       'carheight', 'curbweight', 'cylindernumber', 'enginesize', 'boreratio',
       'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg',
       'highwaympg']])
```

```python
from sklearn.linear_model import Lasso
```

```python
reg = Lasso()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'alpha':  [1e-15,1e-10,1e-5,1e-3,1e-2,1e-1,1,5,10,20,40,50,60,80,100]}
lasso = GridSearchCV(reg, parameters,scoring='r2',cv=5)
lasso.fit(x_train, y_train)
```

```python
print(lasso.best_params_)
print(lasso.best_score_)
```

```python
reg_new = Lasso(alpha=0.01)
reg_new.fit(x_train, y_train)
```

```python
reg_new.coef_
```

```python
y_predict_train = reg_new.predict(x_train)
y_predict_test = reg_new.predict(x_test)
```

```python
from sklearn.metrics import r2_score
train_score = r2_score(y_train, y_predict_train)
test_score = r2_score(y_test, abs(y_predict_test))
```

```python
train_score, test_score
```

```python
from sklearn.metrics import mean_squared_log_error
log_train_score = mean_squared_log_error(y_train, y_predict_train)
log_test_score = mean_squared_log_error(y_test, abs(y_predict_test))
```
